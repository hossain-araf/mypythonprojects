```python
import pandas as pd
from sklearn.tree import DecisionTreeClassifier
df=pd.read_csv('Car_sales.csv')
df.dtypes
dg=df.drop(columns=['Manufacturer','Sales_in_thousands','__year_resale_value','Latest_Launch','Power_perf_factor','Vehicle_type','Price_in_thousands'])
h=dg.drop(columns=['Horsepower','Wheelbase','Width','Length','Curb_weight','Engine_size'])
h.Fuel_capacity=h.Fuel_capacity.astype('float')
h.Fuel_efficiency=h.Fuel_efficiency.astype('float')
x=h.drop(columns=['Model'])
y=h.Model
model=DecisionTreeClassifier()
model.fit(x,y)
car=model.predict([[13,23],[16,23]])
car
```
