```python
import pandas as pd
import seaborn as sns 

sa=pd.read_csv("heart_failure_clinical_records_dataset.csv")
sa.info()
sa.head(12)
dh=sa.drop(columns=['serum_creatinine','serum_sodium'])
dh.info()
dh.age=dh.age.astype(int)
dh.platelets =dh.platelets.astype(int)

dh.info()
from sklearn.tree import DecisionTreeClassifier
x=dh.drop(columns=['DEATH_EVENT'])
y=dh.DEATH_EVENT
hospital=DecisionTreeClassifier()
hospital.fit(x,y)
liveornot=hospital.predict([[],[]])
```
