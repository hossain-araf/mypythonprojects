```python
import pandas as pd
kl=pd.read_csv("Car_sales.csv")
ol=kl.drop(columns=['Manufacturer','Sales_in_thousands','Latest_Launch','__year_resale_value','Power_perf_factor','Curb_weight','Wheelbase','Width','Length'])
io=ol.dropna()#things only needed are selected
io.Engine_size=ol.Engine_size.astype('int')
io.Fuel_capacity=io.Fuel_capacity.astype('int')
io.Price_in_thousands=io.Price_in_thousands.astype("int")#converting float objects to int
io.info()
```

```python
io
```

```python
from sklearn import  preprocessing
le=preprocessing.LabelEncoder()#feature is about the given data
label=le.fit_transform(io.Model)
vehicletype_encoded=le.fit_transform(io.Vehicle_type)#label is for what we are gonna predict
features=list(zip(vehicletype_encoded,io.Price_in_thousands,io.Engine_size,io.Horsepower,io.Fuel_capacity,io.Fuel_efficiency ))
from sklearn.neighbors import KNeighborsClassifier#main prediction
model=KNeighborsClassifier(n_neighbors=12)
model.fit(features,label)
print(model.predict_proba([[1,22,0,123,12,45],[1,22,0,123,12,40]]))
print(model.predict([[1,22,0,123,12,45],[1,22,0,123,12,40]]))
```
