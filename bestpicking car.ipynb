```python
import pandas as pd
from sklearn.tree import DecisionTreeClassifier
from sklearn.model_selection import train_test_split
from sklearn.metrics import accuracy_score
df=pd.read_csv('Car_sales.csv')
df.dtypes
dg=df.drop(columns=['Manufacturer','Sales_in_thousands','__year_resale_value','Latest_Launch','Power_perf_factor','Vehicle_type','Price_in_thousands'])
```

```python
dg.Wheelbase=dg.Wheelbase.astype(int)
dg.Engine_size=dg.Engine_size.astype(int)
dg.Horsepower=dg.Horsepower.astype(int)
dg.Width=dg.Width.astype(int)
dg.Length=dg.Length.astype(int)
dg.Fuel_capacity=dg.Fuel_capacity.astype(int)
dg.Fuel_efficiency=dg.Fuel_efficiency.astype(int)
q=dg.drop(columns=['Curb_weight'])

model=DecisionTreeClassifier()
x=q.drop(columns=['Model'])
y=q.Model
model.fit(x,y)
a=model.predict([[1,225,102,68,123,14,20],[2,223,100,89,120,15,23]])
a
```
